```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=sns.load_dataset('iris')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
filter=dataset['species']=='setosa'
dataset[filter].head()
```

```python
dataset[filter].shape
```

```python
grouped_dataset=dataset.groupby('species').agg({'sepal_length':'mean','sepal_width':'mean','petal_length':'mean','petal_width':'mean'})
grouped_dataset
```

```python
plt.scatter(dataset['petal_length'],dataset['petal_width'],c='red')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('scatter plot')
plt.show()
```

```python
specie=dataset['species'].unique()
vcount=dataset['species'].value_counts()

plt.bar(specie,vcount,color='orange')

plt.title('bar plot')
plt.show()
```

```python
plt.hist(dataset['petal_length'],bins=50,alpha=0.9,color='red',edgecolor='black')
plt.xlabel('petal_length')
plt.title('hist plot')
```

```python
plt.hist(dataset['sepal_width'],bins=50,alpha=0.7,color='blue',edgecolor='black')
```

```python
plt.hist(dataset['petal_width'],bins=50,alpha=0.5,color='green',edgecolor='black')
```

```python
plt.hist(dataset['sepal_length'],bins=50,alpha=0.3,color='orange',edgecolor='black')
```

```python
specie=dataset['species'].unique()
vcount=dataset['species'].value_counts()
plt.pie(vcount,labels=specie,autopct='%1.1f%%')
plt.title('pie plot')
```

```python
sns.pairplot(dataset, hue="species")
```

```python
sns.lineplot(x="petal_length", y="petal_width",hue="species", data=dataset)
```

```python
sns.lineplot(x="sepal_length", y="sepal_width",hue="species",style="species", data=dataset)
```

```python
sns.boxplot(dataset, x="sepal_width", y="species", hue="species",whis=[0, 100], width=.6, palette="vlag")
```

```python
encoded_data = pd.get_dummies(dataset, columns=['species'])
correlation = encoded_data.corr()
sns.heatmap(correlation, annot=True, cmap='viridis')
plt.show()
```
