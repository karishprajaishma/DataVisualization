IMPORTING PACKAGES

```python
import numpy as np
import matplotlib.pyplot as plt
```

SCATTERPLOT 

```python
#DATA FOR X AND Y
x=np.random.rand(50) *100 
y=np.random.rand(50) *100 

#PLOTTING X AND Y WITH COLOUR=BLUE, SIZE=100 ,MARKER(SHAPE) = STAR (*) ALPHA IS TRANSPARENCY OVER OTHER DATAPOINTS  
plt.scatter(x,y,c="b",s=100,marker="*",alpha=0.2)

#TITLE WITH COLOR = YELLOW
plt.title("SCATTER PLOT",c='y')

#XLABEL TO PROVIDE TEXT IN X-AXIS WITH COLOR = RED
plt.xlabel("random value below 100",c='r')

#YLABEL TO PROVIDE TEXT IN Y-AXIS WITH COLOR = BLUE
plt.ylabel("random value below 100",c ='b')
```

LINEPLOT

```python
#CONSIDERING A PERSON'S WEIGHT FOR ABOUT 16 YEARS AS AN EXAMPLE FOR DATA

#YEARS FROM 2006 TO 2021
years=[2006 + x for x in range(16)]
print(years)

#ASSUMPTION OF WEIGHT OF THE PERSON THROUGOUT THE YEARS
weight =[75,74,75,72,74,76,77,74,76,78,75,77,76,76,75,77]

#PLOTTING THE YEARS AND WEIGHTS WITH RED COLOR DASHED LINE BY USING STAR TO REPRESENT THE DATAPOINT AND ADDING LINE THICKNESS IS 1
plt.plot(years,weight,'r--',marker="*",linewidth="1")
```

BARPLOT

```python
#CONSIDER AN EXAMPLE OF IDENTIFYING THE PROGRAM KNOW IN A GROUP

#LIST OF PROGRAMMING LANGUAGES
x=["C++","C#","Python","Java","Go"]

#NO OF PERSONS KNOWING THE PROGRAM
y=[23,10,45,34,23]

#PLOTTING X AS PROG-LANGUAGES AND Y AS THE NO OF PERSONS KNOWS THAT PROG-LANG WITH RED COLOR ,ALIGNMENT FROM EDGE,BAR' WIDTH AS 0.5 ,EDGE COLOR AS BLACK AND LINE WIDTH TO EDGE IS 3
plt.bar(x,y,color='r',align="edge",width=0.5,edgecolor="black",lw=3)
```

HISTOGRAM

```python
#CONSIDER NORMAL DISTRIBUTION WITH AGES FOR EXAMPLE WITH MEAN 50 AND STD DEVIATION AS 1.5 AT BY RANDOM GENERATION FOR GROUP OF 1000 PERSONS

#DATA 
ages=np.random.normal(50,1.5,1000)

#PLOTTING OF THE HISTOGRAM WHERE X-AXIS IS AGES AND Y AXIS IS COUNT(PERSONS) AT SPECIFIC AGE 

#10 BINS(SPLIT INTO 10 BARS OF HISTOGRAMS)

#DENSITY IS FALSE,IF IT IS TRUE THE Y AXIS VALUES IS NORMALISED TO 1 
plt.hist(ages,density=False,bins=10)
```

PIEPLOT

```python
#CONSIDER THE PROGRAMMING LANGS AND VOTES MADE FOR EACH OF THEM 

langs=["Python","C++","Java","C#","Go"]
votes=[65,34,45,32,4]

#HERE, GO LANG IS POPPED OUT AT A SMALL RATE OF 0.2
explodes=[0,0,0,0,0.2]

#PLOTTING LANGUAGES ALONG WITH LABELS AS LANGUAGES ,USING EXPLODES , GENERATING PERCENTAGE FOR INDIVIDUAL VOTES 
# PLACING PERCENTAGE DISTANCE AS 0.8 FROM THE CENTER AND STARTING FROM 90 DEGREE 
plt.pie(votes,labels=langs,explode=explodes,autopct="%.2f%%",pctdistance=0.8,startangle=90)
plt.show()
```

BOXPLOT

```python
#CONSIDER THE NORMAL DISTRIBUTION OF 300 VALUES WITH MEAN HEIGHT OF 172 AND STANDARD DEVIATION OF 8
heights=np.random.normal(172,8,300)

#PLOTTING BOXPLOT IN VERTICAL FORMAT AND ASSIGNING 
# PROPERTIES(COLOR) OF BOXPLOT AS FACECOLOR(INSIDE) AS BLUE AND OUTLINE AS RED
plt.boxplot(heights, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue', color='red'))
```
